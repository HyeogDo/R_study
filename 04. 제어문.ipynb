```r
x <- 10
y <- runif(1) * 100

if (x > y) {
    print("x > runif()")
}
print("---End---")
```

```r
x <- 10
y <- runif(1) * 100

if (x > y) {
    print("x > runif()")
} else {
    print("x <= runif()")
}
print("---End---")
```

```r
x <- 10
y <- runif(1) * 100

if (x > y) {
    print("x > runif()")
} else if(x < y) {
    print("x < runif()")
} else {
    print("x = runif()")
}
print("---End---")
```

```r
# 홀수 짝수 구분

a <- c(8,7,6)
if(a %% 2 == 0) {
    cat(a,"는 짝수", sep="")
    sprintf("%i는 짝수입니다.",a)
} else {
    cat(a,"는 홀수", sep="")
}
```

```r
a <- c(8,7,6)
ifelse(a %% 2 == 0, paste(a,": 짝수"), paste(a,": 홀수"))
```

```r
# 사용자로 부터 data를 입력 받아 입력한 숫자가 홀수 인지 짝수 인지 판단하여 출력하기

num1 <- as.integer(readline("숫자를 입력하세요 : "))

# 입력하신 숫자는 12이고 짝수 입니다.
ifelse(num1 %% 2 == 0, sprintf("입력하신 숫자는 %i이고 짝수입니다.", num1), sprintf("입력하신 숫자는 %i이고 홀수입니다.", num1))
```

```r
# q1컬럼에서 Data가 3보다 크거나 같으면 data의 제곱근을 구하고 아니면 원래 Data를 출력하기

q1 <- excel$q1
ifelse(q1 >= 3, round(sqrt(q1),4), q1)
```

```r
# q1컬럼에서 Data가 3보다 크거나 같으면 data의 제곱근을 구하고 아니면 원래 Data를 q6이라는 컬럼을 만들어 추가하기

q1 <- excel$q1
excel$q6 <- ifelse(q1 >= 3, round(sqrt(q1),4),q1)
head(excel)
```
